```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris_dataset = pd.read_csv("sample_data/iris.csv")
```

```python
X = iris_dataset.iloc[:, :-1]
y = iris_dataset.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)
```

```python
k_values = range(1, 21)
accuracies = []
best_accuracy = 0
best_predictions = []
best_k = 1
```

```python
for k in k_values:
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(X_train, y_train)
    predictions = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    accuracies.append(accuracy)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_predictions = predictions
        best_k = k

# Plotting the accuracy graph
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='green')
plt.title("KNN Accuracy for Different Values of k")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
print(f"\nBest Accuracy: {best_accuracy * 100:.2f}% with k = {best_k}")
print("\n> Predictions vs Actual Values:")
for pred, actual in zip(best_predictions, y_test):
    print(f"> predicted = {pred}, actual = {actual}")
```
